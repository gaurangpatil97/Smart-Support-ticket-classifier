```python
import pandas as pd
train = pd.read_csv('data/train_clean.csv')
test = pd.read_csv('data/test_clean.csv')
```

```python
print(train['label'].unique())
print(test['label'].unique())
# check for spaces, casing issues
```

```python
from langdetect import detect
train['lang'] = train['text'].apply(lambda x: detect(x) if len(x) > 20 else 'unknown')
print(train['lang'].value_counts())
```

```python
import re
def clean_text(text):
    text = re.sub(r'\[.*?\]', '', text)  # remove [NAME], [TICKET ID] etc
    text = re.sub(r'\s+', ' ', text)     # collapse whitespace
    return text.strip().lower()

train['text_clean'] = train['text'].apply(clean_text)
test['text_clean'] = test['text'].apply(clean_text)
```

```python
train = train[train['text_clean'].str.split().str.len() >= 3]
```

```python
train.to_csv('data/train_final.csv', index=False)
test.to_csv('data/test_final.csv', index=False)
```
